## Esercizio finale

Scrivete un breve frammento di codice che:

* legga (usando la funzione [input](https://docs.python.org/3/library/functions.html#input)) il cognome di uno studente , 
* stampi (usando la funzione [print](https://docs.python.org/3/library/functions.html#print)) `AL` se il cognome inizia con una lettera compresa tra `A` e `L`, oppure `MZ` (se il cognome inizia con una lettera compresa tra `M` e `Z`).

### Soluzione

```python
cognome = input("Inserisci il cognome dello studente: ")

if 'A' <= cognome < 'M':
  print('AL')
else:
  print('MZ')
```
