## Conta le vocali

Scrivi un programma che chiede all'utente di inserire una parola (scritta in minuscolo) 
e conta quante vocali ci sono in quella parola.

### Soluzione

La soluzione fa uso di una funzione `conta_vocali` che usa l'operatore `in` per verificare se ogni lettera della parola è una vocale.

```python
VOCALI = 'aeiou'

def conta_vocali(parola):
    count = 0
    for lettera in parola:
        if lettera in VOCALI:
            count += 1
    return count

parola = input("Inserisci una parola: ")
numero_vocali = conta_vocali(parola)
print(f"La parola '{parola}' contiene {numero_vocali} vocali.")
```

## Sequenze di Collatz

Una sequenza di Collatz è una sequenza di numeri interi $c_0, c_1, c_2, \ldots$ definita a partire da un numero intero positivo $c_0$ e seguendo queste regole:

$$
c_{n+1} = \begin{cases}
\frac{c_n}{2} & \text{se } c_n \text{
  è pari} \\  
3c_n + 1 & \text{se } c_n \text{ è dispari}
\end{cases}
$$

Scrivi un programma che, dato un numero intero positivo $c_0$, calcola la sequenza di Collatz a partire da $c_0$ e la stampa fino a quando non raggiunge 1.

```python
def prossimo(n):
  if n % 2 == 0:
    return n // 2
  else:
    return 3 * n + 1

c = int(input("Inserisci un numero: "))
while True:
  print(c)
  c = prossimo(c)
  if c == 1: break
```
