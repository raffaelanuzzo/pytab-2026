```python
def tabelline_n(n):
    moltiplicatore = 1
    tabellina = []
    while moltiplicatore <= 10:
        tabellina.append(n*moltiplicatore)
        moltiplicatore +=1
    return tabellina
```

```python
def firma(parola):
    f= list(parola)
    f.sort()
    return ''.join(f)
```

```python
firma('pazzi')== firma('pizza')
```

```python
parole = ['pizza','pazzi','cane','cena','solitario']
```

```python
firme_a_parole =dict()
for parola in parole:
    f = firma(parola)
    if f in firme_a_parole:
        firme_a_parole[f].append(parola)
    else:
        firme_a_parole[f] = [parola]
for parole in firme_a_parole.values():
    if len(parole) >1:
        print(','.join(parole))
```

```python
from urllib.request import urlopen
```

```python
url = 'https://raw.githubusercontent.com/napolux/paroleitaliane/refs/heads/main/paroleitaliane/60000_parole_italiane.txt'
tutte_le_parole = [p.decode('utf-8').strip() for p in urlopen(url)]
tutte_le_parole
len(tutte_le_parole)
```

```python
firme_a_parole =dict()
for parola in tutte_le_parole:
    f = firma(parola)
    if f in firme_a_parole:
        firme_a_parole[f].append(parola)
    else:
        firme_a_parole[f] = [parola]
for parole in firme_a_parole.values():
    if len(parole) >1:
        print(','.join(parole))
```

```python
decorato = []
for parole in firme_a_parole.values():
    decorato.append((len(parole),parole))
decorato.sort(reverse=True)    
```
