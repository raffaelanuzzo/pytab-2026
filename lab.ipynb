##CONTA LE VOCALI
Scrivere un programa che chiede una stringa (in minuscolo) in input e stampra il numero di vocali che contiene 

```python
vocali = ['a','e','i','o','u','è','é','ò','ù']
counter_vocali = 0
parola = str(input()).lower()
for lettera in parola:
    if lettera in vocali:
        counter_vocali +=1
print(f'Il numero di vocali è {counter_vocali}')
```

```python
## Sequenze di Collatz Se n è pari, passo a n/2, altrimento passo 3*n +1 
```

```python
numero = int(input())
while True:
    if numero == 1:
        print(f'Il numero inserito è pari a 1')
        break
    elif numero % 2 == 0:
         numero = numero // 2
         print(numero)
    else:
        numero = 3*numero + 1
        print(numero)

```

Funzioni: posso usare più volte lo stsso codice

```python
def dieci_ciao():
    for i in range(10):
        print('Ciao')
dieci_ciao()
print('E ancora')
dieci_ciao()
```

```python
def dieci_ciao(saluto):
    for i in range(10):
        print(saluto)
dieci_ciao('ciao')
print('e ancora')
dieci_ciao('arrivederci')
```

```python
def saluti(saluto, nvolte = 4, spazi =2): #parametri di funzione, specificato il default
    """
    la funzione stampa nvolte il saluto:
    Parametri:
    - saluto tipo stringa
    - nvolte tipo in
    """
    for i in range(nvolte):
        print(spazi * ' ' + saluto)
saluti('ciao')
```

```python
help(saluti)
```

Le funzioni possono avere n parametri, se vogliamo usare il valore di default non è neseccario inserirlo in invocazione. I parametri vengono gestini in ordine, seguendo la posizione di definizione. Si possono utilizzare le definizioni del parametro per invocare la funzione se non ricordo l'ordine, ma da quel punt in poi devo usarli per tutit i parametri rimanenti:
saluti('ciao', spazi = 5, nvolte = 2) OK
saluti(nvolte = 2, 'CIAO', spazi = 4)  KO -> DEVO INSERIRE IL NOME DEL PAMETRO

```python
def conta_vocali(testo):
    n=0
    for lettera in testo:
        if lettera in 'aeiou': n+=1
    return n
```

```python
conta_vocali( 'ciao nonna') == conta_vocali('ciao') + conta_vocali('nonna')
```

```python
n=0
for linea in urlopen('https://www.repubblica.it'):
    n+= conta_vocali(linea.decode('utf8'))
print(n)
```
